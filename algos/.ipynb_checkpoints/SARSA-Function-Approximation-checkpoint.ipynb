```python
%matplotlib inline

import gym
import itertools
import matplotlib
import numpy as np
import pandas as pd
import sys

if "../" not in sys.path:
  sys.path.append("../") 

from collections import defaultdict
from lib.envs.cliff_walking import CliffWalkingEnv
from lib import plotting
import math

matplotlib.style.use('ggplot')
```

```python
env = gym.make("CartPole-v0")
```

```python
class QFunction:
    """
    2 layer vanilla neural network with sigmoid activation functions implemented entirely in numpy 
    
    """
    def __init__(self,action_space,observ_dim=1,K=10,seed=1):
        np.random.seed(1)
        self.observ_dim = observ_dim
        self.w1 = 2 * np.random.random((self.observ_dim, K)) - 1 
        self.w2 = 2 * np.random.random((K, action_space)) - 1
        self.l0 = np.zeros(1)
        self.l1 = np.zeros(1)
        self.l2 = np.zeros(1)
        self.action_space = action_space
        self.observ_space = self.observ_dim
    
    def sigmoid_derivative(self, X):
        return X * (1 - X)
    
    def sigmoid(self, X): 
        return 1 / (1 + np.exp(-X))
    
    def forward_pass(self, observation):
        l0 = observation.T
        l1 = self.sigmoid(np.dot(l0.T, self.w1))
        l2 = self.sigmoid(np.dot(l1, self.w2))
        return l0, l1, l2
    
    def backprop(self, target, observation, alpha):        
        l2_error = self.l2 - target
        l2_delta = l2_error * self.sigmoid_derivative(self.l2)
        l1_error = np.dot(l2_delta, self.w2.T)
        l1_delta = l1_error * self.sigmoid_derivative(self.l1)
        self.w2 -= alpha * np.dot(self.l1.T, l2_delta)
        self.w1 -= alpha * np.dot(self.l0, l1_delta) 
        
    
    def set_layer_outputs(self, l0, l1, l2):
        self.l0 = l0
        self.l1 = l1 
        self.l2 = l2
    

```

```python
def get_action(env, Q_values, epsilon):
        probs = np.ones(env.action_space.n, dtype=float) * epsilon / env.action_space.n
        best_action = np.argmax(Q_values)
        probs[best_action] += (1.0 - epsilon)
        action = np.random.choice(np.arange(env.action_space.n), p=probs)
        return action

MIN_EXPLORE_RATE  = 0.01 
MIN_LEARNING_RATE = 0.1

def get_explore_rate(t):
    return max(MIN_EXPLORE_RATE, min(1, 1.0 - math.log10((t+1)/25)))

def get_learning_rate(t):
    return max(MIN_LEARNING_RATE, min(0.5, 1.0 - math.log10((t+1)/25)))    
    

def sarsa(env, num_episodes, discount_factor=1.0, alpha=0.5, epsilon=0.1):
    """
    SARSA algorithm with neural network function approximation
    
    """
    
    # The final action-value function.
    # A nested dictionary that maps state -> (action -> action-value).
    Q_func = QFunction(env.action_space.n, observ_dim=4)
    # Keeps track of useful statistics
    episode_lengths= []
    episode_rewards= [] 
    
    
    for i_episode in range(num_episodes):
        # Print out which episode we're on, useful for debugging.
        epsilon = get_explore_rate(float(i_episode))
        alpha = get_learning_rate(float(i_episode))
        if (i_episode + 1) % 100 == 0:
            print "\rEpisode {}/{}.".format(i_episode + 1, num_episodes)
            sys.stdout.flush()
        
        done = False
        
        s = env.reset()
        
        if Q_func.observ_space == 1:
            s = np.array([[s]])
        else:
            s = np.array([s])
        
        episode_l = 0
        episode_r = 0
        l0, l1, l2 = Q_func.forward_pass(s)
        
        Q_values = l2.T[: , 0]
        
        action = get_action(env, Q_values, epsilon)
        
        Q_func.set_layer_outputs(l0, l1, l2)
        
        while not done:
           
            next_s, r, done, _ = env.step(action)
            if Q_func.observ_space == 1:
                next_s = np.array([[next_s]])
            else:
                next_s = np.array([next_s])
            
            
            if done:
                episode_l += 1
                episode_r += r
                Q_func.backprop(r, s, alpha)
                break
            
            l0, l1, l2 = Q_func.forward_pass(next_s)
            
            next_Q_values = l2.T[: , 0]
            
            next_action = get_action(env, next_Q_values, epsilon)
             
            target = r + discount_factor * next_Q_values[next_action]
            
            Q_func.backprop(target, s, alpha)
            
            l0, l1, l2 = Q_func.forward_pass(next_s)
            
            Q_func.set_layer_outputs(l0, l1, l2)
            
            action = next_action
            
            s = next_s
            episode_l += 1
            episode_r += r

        episode_r /= float(episode_l)
        episode_lengths.append(episode_l)
        episode_rewards.append(episode_r)
    stats = plotting.EpisodeStats(episode_lengths=np.array(episode_lengths), episode_rewards=np.array(episode_rewards))

    return stats
```

```python
stats = sarsa(env, 10000)
```

```python
plotting.plot_episode_stats(stats)
```
